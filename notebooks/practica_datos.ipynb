Agregue una columna tipo date formada por los valores de las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.

```python

import numpy as np

# cargamos el DataFrame
df = pd.read_csv('..\\data\\atus_combinado.csv')

# convertimos columnas a numérico, reemplazando valores no válidos con NaN
for col in ['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# una Función para crear fecha con manejo de valores nulos
def crear_fecha(row):
    try:
        return pd.to_datetime(
            f"{int(row['ANIO'])}-{int(row['MES'])}-{int(row['ID_DIA'])} {int(row['ID_HORA']):02d}:{int(row['ID_MINUTO']):02d}",
            format='%Y-%m-%d %H:%M'
        )
    except:
        return pd.NaT

# creamos la columna de fecha
df['FECHA'] = df.apply(crear_fecha, axis=1)

# mostramos las primeras filas del DataFrame
print("\nPrimeras filas del DataFrame con la nueva columna FECHA:")
print(df[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO', 'FECHA']].head(10))

# Información sobre la nueva columna FECHA
print("\nInformación sobre la nueva columna FECHA:")
print(df['FECHA'].describe())

# Verificar si hay  fechas nulas
fechas_nulas = df[df['FECHA'].isnull()]
print(f"\nNúmero de fechas nulas: {len(fechas_nulas)}")

if not fechas_nulas.empty:
    print("\nEjemplos de filas con fechas nulas:")
    print(fechas_nulas[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']].head())

# Estadísticas de valores nulos en las columnas originales
print("\nEstadísticas de valores nulos en las columnas originales:")
for col in ['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']:
    print(f"{col}: {df[col].isnull().sum()} valores nulos")

# Opcionalmente, guardar el DataFrame actualizado
#df.to_csv('..\\data\\atus_combinado_con_fecha.csv', index=False)

print("\nLa columna FECHA ha sido agregada al DataFrame original.")
```

Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.

```python



# Cargar el DataFrame
df = pd.read_csv('..\\data\\atus_combinado.csv')

# Convertir columnas a numérico, reemplazando valores no válidos con NaN
for col in ['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO', 'URBANA', 'SUBURBANA']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Función para crear fecha con manejo de valores nulos
def crear_fecha(row):
    try:
        return pd.to_datetime(
            f"{int(row['ANIO'])}-{int(row['MES'])}-{int(row['ID_DIA'])} {int(row['ID_HORA']):02d}:{int(row['ID_MINUTO']):02d}",
            format='%Y-%m-%d %H:%M'
        )
    except:
        return pd.NaT

# Crear la columna de fecha
df['FECHA'] = df.apply(crear_fecha, axis=1)

# Función para determinar el área
def determinar_area(row):
    if row['URBANA'] == 1:
        return 'Urbana'
    elif row['SUBURBANA'] == 1:
        return 'Suburbana'
    else:
        return 'No especificada'

# Crear la columna de Área
df['AREA'] = df.apply(determinar_area, axis=1)

# Mostrar las primeras filas del DataFrame
print("\nPrimeras filas del DataFrame con las nuevas columnas FECHA y AREA:")
print(df[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO', 'FECHA', 'URBANA', 'SUBURBANA', 'AREA']].head(10))

# Información sobre la nueva columna FECHA
print("\nInformación sobre la nueva columna FECHA:")
print(df['FECHA'].describe())

# Información sobre la nueva columna AREA
print("\nDistribución de valores en la columna AREA:")
print(df['AREA'].value_counts(dropna=False))

# Verificar fechas nulas
fechas_nulas = df[df['FECHA'].isnull()]
print(f"\nNúmero de fechas nulas: {len(fechas_nulas)}")

if not fechas_nulas.empty:
    print("\nEjemplos de filas con fechas nulas:")
    print(fechas_nulas[['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO']].head())

# Estadísticas de valores nulos en las columnas originales
print("\nEstadísticas de valores nulos en las columnas originales:")
for col in ['ANIO', 'MES', 'ID_DIA', 'ID_HORA', 'ID_MINUTO', 'URBANA', 'SUBURBANA']:
    print(f"{col}: {df[col].isnull().sum()} valores nulos")

# Opcionalmente, guardar el DataFrame actualizado
#df.to_csv('..\\data\\atus_combinado_con_fecha_y_area.csv', index=False)

print("\nLas columnas FECHA y AREA han sido agregadas al DataFrame original.")
```

Determine cual municipio de cada estado tuvo más meses sin accidentes

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el DataFrame
df = pd.read_csv('..\\data\\atus_combinado_final.csv')

# 1. Municipio con más meses sin accidentes por estado
def meses_sin_accidentes(grupo):
    todos_meses = set(range(1, 13))
    meses_con_accidentes = set(grupo['MES'].unique())
    return len(todos_meses - meses_con_accidentes)

municipios_sin_accidentes = df.groupby(['ID_ENTIDAD', 'ID_MUNICIPIO']).apply(meses_sin_accidentes).reset_index()
municipios_sin_accidentes.columns = ['ID_ENTIDAD', 'ID_MUNICIPIO', 'MESES_SIN_ACCIDENTES']
municipios_max_sin_accidentes = municipios_sin_accidentes.loc[municipios_sin_accidentes.groupby('ID_ENTIDAD')['MESES_SIN_ACCIDENTES'].idxmax()]

print("1. Municipios con más meses sin accidentes por estado:")
print(municipios_max_sin_accidentes)
```

Resuma los accidentes del último año por tipo y que porcentaje representan

```python
#  Resumen de accidentes del último año por tipo
ultimo_anio = df['ANIO'].max()
accidentes_ultimo_anio = df[df['ANIO'] == ultimo_anio]['TIPACCID'].value_counts()
porcentaje_accidentes = (accidentes_ultimo_anio / accidentes_ultimo_anio.sum() * 100).round(2)

print("\n2. Resumen de accidentes del último año por tipo:")
print(pd.DataFrame({'Cantidad': accidentes_ultimo_anio, 'Porcentaje': porcentaje_accidentes}))
```

Obtenga la cantidad de muertos y lesionados de los últimos 3 años

```python
# Función para convertir columnas a numérico, manejando errores
def to_numeric_safe(df, column):
    df[column] = pd.to_numeric(df[column], errors='coerce')
    return df

# Lista de columnas a convertir
columnas_numericas = ['CONDMUERTO', 'CONDHERIDO', 'PASAMUERTO', 'PASAHERIDO',
                      'PEATMUERTO', 'PEATHERIDO', 'CICLMUERTO', 'CICLHERIDO',
                      'OTROMUERTO', 'OTROHERIDO', 'NEMUERTO', 'NEHERIDO']

# Aplicar la conversión a todas las columnas necesarias
for col in columnas_numericas:
    df = to_numeric_safe(df, col)

#  Cantidad de muertos y lesionados de los últimos 3 años
ultimos_3_anios = df['ANIO'].unique()[-3:]
muertos_lesionados = df[df['ANIO'].isin(ultimos_3_anios)].groupby('ANIO').agg({
    col: 'sum' for col in columnas_numericas
})

muertos_lesionados['Total_Muertos'] = muertos_lesionados[[col for col in muertos_lesionados.columns if 'MUERTO' in col]].sum(axis=1)
muertos_lesionados['Total_Heridos'] = muertos_lesionados[[col for col in muertos_lesionados.columns if 'HERIDO' in col]].sum(axis=1)

print("\n3. Cantidad de muertos y lesionados de los últimos 3 años:")
print(muertos_lesionados[['Total_Muertos', 'Total_Heridos']])

# Verificar si hay valores nulos después de la conversión
nulos_despues_conversion = df[columnas_numericas].isnull().sum()
print("\nValores nulos después de la conversión:")
print(nulos_despues_conversion)

# Si hay valores nulos, mostrar algunas filas de ejemplo
if nulos_despues_conversion.sum() > 0:
    print("\nEjemplos de filas con valores no numéricos:")
    filas_problematicas = df[df[columnas_numericas].isnull().any(axis=1)]
    print(filas_problematicas[['ANIO'] + columnas_numericas].head())
```

¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?

```python
#4. Comparación del número de accidentes por año
accidentes_por_anio = df.groupby('ANIO').size()
variacion_porcentual = accidentes_por_anio.pct_change() * 100

print("\n4. Comparación del número de accidentes por año y variación porcentual:")
print(pd.DataFrame({'Número de Accidentes': accidentes_por_anio, 'Variación Porcentual': variacion_porcentual}))
```
